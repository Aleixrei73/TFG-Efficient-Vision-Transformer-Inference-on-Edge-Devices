```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
smallDataTrain = pd.read_csv('summary/ViT-Distillation/TrainingData/SmallTrain.csv')
smallDataTest = pd.read_csv('summary/ViT-Distillation/TrainingData/SmallTest.csv')
```

```python
plt.plot(smallDataTrain["Value"]);
plt.plot(smallDataTest["Value"]);
plt.legend(["Train accuracy", "Test accuracy"]);
plt.title("DeiT-Small training accuracies")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.savefig("summary/ViT-Distillation/TrainingData/graphs/Small.png")
```

```python
regDataTrain1 = pd.read_csv('summary/ViT-Distillation/TrainingData/RegularizedTrain1.csv')
regDataTest1 = pd.read_csv('summary/ViT-Distillation/TrainingData/RegularizedTest1.csv')
regDataTrain2 = pd.read_csv('summary/ViT-Distillation/TrainingData/RegularizedTrain2.csv')
regDataTest2 = pd.read_csv('summary/ViT-Distillation/TrainingData/RegularizedTest2.csv')
```

```python
train_vals = regDataTrain1["Value"].to_numpy()
train_vals = np.append(train_vals, regDataTrain2["Value"].to_numpy())
test_vals = regDataTest1["Value"].to_numpy()
test_vals = np.append(test_vals, regDataTest2["Value"].to_numpy())
```

```python
plt.plot(train_vals);
plt.plot(test_vals);
plt.legend(["Test accuracy", "Train accuracy"]);
plt.title("Training accuracies")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.savefig("summary/ViT-Distillation/TrainingData/graphs/Regularized.png")
```
